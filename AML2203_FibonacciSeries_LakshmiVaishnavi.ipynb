**Program to print fibonacci sequence using recursion** 

```python
import time
import matplotlib.pyplot as plt
N = int( input("enter the number"))
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
    
#Fibonacci_list=[]
#Time_list=[] 
def generate_fibonacci_numbers(n):
    times = []
    numbers = []
    for i in range(n+1):
        start = time.time()
       # Fibonacci_list.append(fibonacci(n))
        number = fibonacci(i)
        end = time.time()
        numbers.append(number)
        times.append(end-start)
    return numbers, times
print("Fibonacci Series:\n", numbers)
print("Time taken to generate each Fibonacci number:\n", times)

def plot_fibonacci(numbers, times):
    #fig, ax = plt.subplots()
    plt.plot(numbers, times, 'o-')
    plt.plot(xlabel='Fibonacci Number', ylabel='Time (seconds)',
           title='Fibonacci Numbers vs. Time')
    plt.show()

numbers, times = generate_fibonacci_numbers(N)
plot_fibonacci(numbers, times)
```
